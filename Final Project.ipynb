```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
data
```

```python
df_summary = pd.DataFrame(data[0]['matchSummary'])
df_summary.head()
```

```python
df_summary.rename({'scorecard':'match_id'},axis = 1, inplace = True)
df_summary.head()
```

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

In order to understand which players are out/ not out, we would need to look at the dismissal column where there is no value which means the player is not out

So the logic here is to create a new column 'out/not_out' and then use the APPLY function to transform the data on the dissimal column

Creating a new column in python

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")

df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(11)
```

```python
print(df_batting['match'].head(50))
```

```python
Match_ids_dict = {}

for index, row in df_summary.iterrows():
    key1 = row["team1"] + " VS " + row["team2"]
    key2 = row["team2"] + " VS " + row["team1"]
    
    Match_ids_dict[key1] = row["match_id"]
    Match_ids_dict[key2] = row["match_id"]
    
Match_ids_dict
```

```python
Match_ids_dict["Bangladesh VS Pakistan"]
```

```python
df_batting['Maatch_id'] = df_batting['match'].map(Match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index= False)
```
